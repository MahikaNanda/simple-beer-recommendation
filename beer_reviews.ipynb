```python
beers = pd.read_csv('beer_reviews.csv')
```

```python
beers1 = beers.ix[:(beers["brewery_id"].count())/2,:]
```

```python
beers1.describe()
```

```python
beers1.head()
```

```python
my_beers = pd.DataFrame()
my_beers = beers1[["review_overall","review_appearance","review_aroma","beer_style","review_taste","beer_name","beer_abv","review_profilename"]]
```

```python
my_beers.head()
```

```python
my_beers.describe()
```

```python

avg_overall_rating= pd.DataFrame(my_beers.groupby('beer_name')['review_overall'].mean())
avg_overall_rating.head()
```

```python
avg_overall_rating['no_of_rating'] = my_beers.groupby('beer_name')['review_overall'].count()
avg_overall_rating.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
avg_overall_rating['review_overall'].hist(bins=25)
```

```python
avg_overall_rating['no_of_rating'].hist(bins=10)
```

```python
import seaborn as sns
sns.jointplot(x='review_overall', y='no_of_rating', data=avg_overall_rating)
```

```python
beer_matrix = my_beers.pivot_table(index = 'review_profilename',columns= 'beer_name',values = 'review_overall')
beer_matrix.head(10)
```

```python
avg_overall_rating.sort_values('no_of_rating', ascending=False).head(10)
```

```python
sierra_nevada_paleale_user_rating = beer_matrix['Sierra Nevada Pale Ale']
breakfast_stout_user_rating = beer_matrix['Founders Breakfast Stout']
sierra_nevada_paleale_user_rating.head(20)
```

```python
breakfast_stout_user_rating.head(20)
```

```python
similiar_to_breakfast_stout = beer_matrix.corrwith(breakfast_stout_user_rating)
similiar_to_breakfast_stout.head(25)
```

```python
similiar_to_sierra_nevada_paleale_user_rating = beer_matrix.corrwith(sierra_nevada_paleale_user_rating)
similiar_to_sierra_nevada_paleale_user_rating.head(25)
```

```python
corr_sierra_nevada_paleale = pd.DataFrame(similiar_to_sierra_nevada_paleale_user_rating, columns = ['correlation'])
corr_sierra_nevada_paleale.dropna(inplace =True)
corr_sierra_nevada_paleale.head()
```

```python
corr_breakfast_stout = pd.DataFrame(similiar_to_breakfast_stout, columns = ['correlation'])
corr_breakfast_stout.dropna(inplace = True)
corr_breakfast_stout.head(15)
```

```python
corr_breakfast_stout = corr_breakfast_stout.join(avg_overall_rating['no_of_rating'])
corr_sierra_nevada_paleale = corr_sierra_nevada_paleale.join(avg_overall_rating['no_of_rating'])
corr_breakfast_stout.head()
```

```python
corr_sierra_nevada_paleale.head()
```

```python
corr_breakfast_stout[corr_breakfast_stout['no_of_rating']> 100].sort_values(by = 'correlation',ascending =False).head(15)
```
